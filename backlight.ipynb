# Operasi foto Backlight

## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Membaca Gambar

```python
rgb = cv2.imread('backlight.jpeg')
rgb = cv2.cvtColor(rgb, cv2.COLOR_BGR2RGB)
```

## Mengubah ke Grayscale

```python
gray = cv2.cvtColor(rgb, cv2.COLOR_BGR2GRAY)
baris, kolom = gray.shape
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].imshow(rgb)
axs[0].set_title("Citra RGB")
axs[0].axis("off")

axs[1].imshow(gray, cmap='gray')
axs[1].set_title("Citra Grayscale")
axs[1].axis("off")

plt.tight_layout()
plt.show()
```

## Peningkatan Kecerahan

```python
beta = 40
citra_cerah = np.zeros((baris, kolom), dtype=np.float32)

for x in range(baris):
    for y in range(kolom):
        gyx = gray[x, y] + beta
        citra_cerah[x, y] = gyx

citra_cerah = np.clip(citra_cerah, 0, 255).astype(np.uint8)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(gray, cmap='gray')
axs[0].set_title("Grayscale Asli")
axs[0].axis('off')
axs[1].imshow(citra_cerah, cmap='gray')
axs[1].set_title("Peningkatan Kecerahan")
axs[1].axis('off')
plt.tight_layout()
plt.show()
```

## Peregangan Kontras


```python
alpha = 1.2
citra_kontras = np.zeros((baris, kolom), dtype=np.float32)

for x in range(baris):
    for y in range(kolom):
        gyx = gray[x, y] * alpha
        citra_kontras[x, y] = gyx

citra_kontras = np.clip(citra_kontras, 0, 255).astype(np.uint8)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(gray, cmap='gray')
axs[0].set_title("Grayscale Asli")
axs[0].axis('off')
axs[1].imshow(citra_kontras, cmap='gray')
axs[1].set_title("Peregangan Kontras")
axs[1].axis('off')
plt.tight_layout()
plt.show()
```

## Gabungan Kecerahan + Kontras


```python
alpha = 1.4
beta = 50
citra_hasil = np.zeros((baris, kolom), dtype=np.float32)

for x in range(baris):
    for y in range(kolom):
        gyx = alpha * gray[x, y] + beta
        citra_hasil[x, y] = gyx

citra_hasil = np.clip(citra_hasil, 0, 255).astype(np.uint8)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(gray, cmap='gray')
axs[0].set_title("Grayscale Asli")
axs[0].axis('off')
axs[1].imshow(citra_hasil, cmap='gray')
axs[1].set_title("Gabungan Cerah + Kontras")
axs[1].axis('off')
plt.tight_layout()
plt.show()
```
