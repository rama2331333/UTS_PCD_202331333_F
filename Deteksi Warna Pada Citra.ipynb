# Deteksi Warna Pada Citra

## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Membaca Citra

```python
image = cv2.imread('cek_rgb.jpeg')
```

## Mengkonversi dari BGR ke RGB

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
baris, kolom, _ = image_rgb.shape
```

## Mengkontraskan Citra

```python
alpha = 1.2
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = image_rgb[x, y] * alpha
        citra_kontras[x, y] = np.clip(gyx, 0, 255)  
        
citra_kontras = citra_kontras.astype(np.uint8)
```

## Split Warna

```python
blue_channel = citra_kontras[:, :, 2]
green_channel = citra_kontras[:, :, 1]
red_channel = citra_kontras[:, :, 0]
```

## Menampilkan Hasil

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("Hasil Deteksi Warna")

# Original Image with Contrast
axs[0, 0].imshow(citra_kontras)
axs[0, 0].set_title("Citra Kontras")

# Blue Channel
axs[0, 1].imshow(blue_channel, cmap="gray")
axs[0, 1].set_title("Biru")

# Red Channel
axs[1, 0].imshow(red_channel, cmap="gray")
axs[1, 0].set_title("Merah")

# Green Channel
axs[1, 1].imshow(green_channel, cmap="gray")
axs[1, 1].set_title("Hijau")

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

## Menampilkan Histogram

```python
fig, axs = plt.subplots(4, 1, figsize=(12, 16))
fig.suptitle("Histogram Warna")

axs[0].hist(citra_kontras.ravel(), 256, [0, 256], color='gray')
axs[0].set_title("Histogram Citra Kontras")

axs[1].hist(blue_channel.ravel(), 256, [0, 256], color='blue')
axs[1].set_title("Histogram Biru")

axs[2].hist(red_channel.ravel(), 256, [0, 256], color='red')
axs[2].set_title("Histogram Merah")

axs[3].hist(green_channel.ravel(), 256, [0, 256], color='green')
axs[3].set_title("Histogram Hijau")

plt.tight_layout()
plt.subplots_adjust(top=0.95)
plt.show()
```
